```python
from sklearn.datasets import load_iris
from sklearn.feature_selection import SelectKBest, chi2
X, y = load_iris().data,load_iris().target

X.shape #取得x的維度
kbest = SelectKBest(chi2, k=3)
X_new = kbest.fit_transform(X, y)
X_new.shape
```

```python
kbest.get_params()
```

```python
kbest.get_support() #得分最低 = Col2
```

```python
kbest.inverse_transform(X_new) #Col2資料全部歸0
```

```python
kbest = SelectKBest(chi2, k=1)
X_new = kbest.fit_transform(X, y)
X_new.shape
```

```python
kbest.get_support() #得分最高 = Col3
```

```python
kbest.inverse_transform(X_new) #除Col3外的資料全部歸0
```

```python
kbest.scores_ #各column score
```
