```python
import pandas as pd

data = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')
data.columns = ['Person ID', 'Gender', 'Age', 'Occupation', 'Sleep Duration', 'Quality of Sleep', 'Physical Activity Level', 'Stress Level', 'BMI Category', 'Blood Pressure', 'Heart Rate', 'Daily Steps', 'Sleep Disorder']
data = data.drop(['Person ID'], axis=1)
print('Number of instances = %d' % (data.shape[0]))
print('Number of attributes = %d' % (data.shape[1]))
data.head()
```

```python
from sklearn.model_selection import train_test_split
x = data.loc[:,['Gender', 'Occupation', 'Quality of Sleep','Stress Level']]
x
```

```python
from sklearn.preprocessing import MinMaxScaler
z = data.loc[:,['Age', 'Sleep Duration', 'Physical Activity Level']]
scaler = MinMaxScaler(feature_range = (0,1))
print(scaler.fit(z))
print(scaler.transform(z))
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(z)
print(scaler.transform(z))
```

```python
import numpy as np
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
pca.fit(z)
print(pca.transform(z))
```

```python
from sklearn.feature_selection import VarianceThreshold
selector = VarianceThreshold(threshold = 5)
selector.fit(z)
selector.get_support()
```

```python
print(selector.fit_transform(z))
```

```python
from sklearn.datasets import load_iris
from sklearn.feature_selection import SelectKBest, chi2
w = load_iris().target
kbest = SelectKBest(chi2, k=2)
z_train, z_test = train_test_split(z,test_size=150,random_state=2)
z_new = kbest.fit_transform(z_test, w)
z_new.shape
```

```python
kbest.get_support() #檢視哪一欄的得分最低，將被剔除
```

```python
kbest.inverse_transform(z_new) #col2 data全部歸零
```
